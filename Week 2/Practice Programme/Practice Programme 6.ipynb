Check if a number is divisible by 5, 7 and 9 as well.

```python
def isdivi(num):
  if(num%5==0):
    if(num%7==0):
      if(num%9==0):
        print('Number is divisible by 5,7 and 9')
      else:
        print('Number is divisible by 5 and 7 but not divisible by 9')
    else:
      print('Number is divisible by 5 but not divisible by 7 and 9')
  elif(num%7==0):
    if(num%9==0):
      print('Number is divisible by 7 and 9 but not divisible by 5')
    else:
      print('Number is divisible by 7 but not divisible by 5 or 9')
  elif(num%9==0):
    print('Number is divisible by 9 but not divisible by 7 or 5')
  else:
    Print('Number is not divisible by 5 nor 7 nor 9')
```

## Perform EDA on the dataset that consists of placement data of students in XYZ campus.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Placement_Data_Full_Class.csv')
df
```

#### Lets convert all the categorical data into numerical type so that we can easily analyse which features affects being placed.

```python
#1) ssc_b

ssc_bn = []

for i in df['ssc_b']:
    if (i == df['ssc_b'].unique()[0]):
        ssc_bn.append(0)                 #0 for others
    else:
        ssc_bn.append(1)                 #1 for central

df['ssc_b'] = ssc_bn
```

```python
#2) hsc_b

hsc_bn = []

for i in df['hsc_b']:
    if (i == df['hsc_b'].unique()[0]):
        hsc_bn.append(0)                #0 for others
    else:
        hsc_bn.append(1)                #1 for central

df['hsc_b'] = hsc_bn 
```

```python
#3) hsc_s

hsc_sn = []

for i in df['hsc_s']:
    if (i == df['hsc_s'].unique()[2]):
        hsc_sn.append(0)                  #0 for arts
    elif (i == df['hsc_s'].unique()[0]):
        hsc_sn.append(1)                  #1 for commerce
    else:
        hsc_sn.append(2)                  #2 for science

df['hsc_s'] = hsc_sn
```

```python
#4) degree_t

degree_tn = []

for i in df['degree_t']:
    if (i == df['degree_t'].unique()[2]):
        degree_tn.append(0)                 #0 for others
    elif (i == df['degree_t'].unique()[1]):
        degree_tn.append(1)                 #1 for Comm&Mgmt
    else:
        degree_tn.append(2)                 #2 for Sci&Tech

df['degree_t'] = degree_tn
```

```python
#5) workex

workex_n = []
for i in df['workex']:
    if (i == df['workex'].unique()[0]):
        workex_n.append(0)                 #0 for No
    else:
        workex_n.append(1)                 #1 for Yes

df['workex'] = workex_n
```

```python
#6) Specialisation

spe = []

for i in df['specialisation']:
    if (i == df['specialisation'].unique()[0]):
        spe.append(0)                              #0 for Mkt&HR
    else:
        spe.append(1)                              #1 for Mkt&Fin

df['specialisation'] = spe
```

```python
#7) status

status_n = []

for i in df['status']:
    if (i == df['status'].unique()[1]):
        status_n.append(0)                #0 for Not Placed
    else:
        status_n.append(1)                #1 for placed

df['status'] = status_n
```

```python
df
```

#### So after changing all the categorical values into numerical values our data looks something like above. Now we can move further with our analysis.

### 1) Which factor influenced a candidate in getting placed.

#### To get the ans to this question I will first find the correlation matrix of this data and then rank all factors that affects the candidate in getting placed.

```python
# Correlation Matrix
df_corr = df.corr()
df_corr.style.background_gradient( cmap = 'BrBG')
```

```python
#Ranking the factors
df_rank = df_corr['status'].sort_values(ascending = False)
print(df_rank)
```

#### Above are the factors ranked in descending order of their impact on a candidate getting placed.  

### 2) Does percentage matters for one to get placed?

#### To get this answer lets look at the factors affecting the candidate getting placed once again..

```python
df_rank
```

#### Here we can observe that the top 3 factors are the candidate's percentage that affects their placements. So YES percentage matters for one to get placed.

### 3) Which degree specialisation is much demanded by corporate

#### Lets find the distribution of the specialisation column that got placed to find out this answer..

```python
MktnHR  = 0
MktnFin = 0

for i in df.values:
    if (i[-2] == 1):            # if the candidate is placed
        if (i[-4] == 0):        # if specialisation is MktnHR
            MktnHR += 1
        else:                   # else Specialisation is MktnFin
            MktnFin += 1
```

```python
plt.bar( ['Mkt&HR', 'Mkt&Fin'] , [MktnHR, MktnFin] )
plt.title('Specialisation that is much demanded by the corporate')
plt.xlabel('Specialisation')
plt.ylabel('NO. of candidate that got placed')
plt.show()
```

#### From the above graph we can conclude that Marketing and Finance specialisation is much demanded by corporate.

### 5) Explore the relationshhip between columns using correlation matrix.

#### 5.1) Correlation matrix of Math dataframe.

```python
import seaborn
import matplotlib.pyplot as plt
```

```python
correlation_mat = df_mat.corr()  
correlation_mat.style.background_gradient (cmap = 'BrBG')
```

```python
fig, ax = plt.subplots(figsize = (14,14))
heatmap = seaborn.heatmap(correlation_mat, annot = True, ax = ax)
heatmap.set(title = 'Correlation Matrix with heatmap of Math DataFrame')
plt.show()
```

#### 5.2) Correlation matrix of the Portugese Dataframe

```python
correlation_por = df_por.corr()
correlation_por.style.background_gradient(cmap= 'BrBG')
```

```python
fig , ax = plt.subplots(figsize = (14,14))
heatmap = seaborn.heatmap(correlation_por, annot = True, ax = ax)
heatmap.set(title = 'Correlation matrix with heatmap of portugese dataframe')
plt.show()
```

### 6) Rank all the features that would affect overall Grade

#### 6.1) Ranking all the features for the maths dataset.

```python
corr_mat_matrix  = df_mat.corr()
corr_mat_ranking = corr_mat_matrix['G3'].sort_values(ascending = False)
print(corr_mat_ranking)
```

Above are the features ranked that affects the overall grade.
The features with a negative float value has an negative correlation and features with no negative float value has an positive correlation. 
That simply means better marks in the first period and second period examination as well as more studytime will result in better final results. 
And
Lesser alcohol consumption and less travelling time will result in better final results.

#### 6.2) Ranking all the features for the portugese dataset.

```python
corr_por_matrix  = df_por.corr()
corr_por_ranking = corr_por_matrix['G3'].sort_values(ascending = False)
print(corr_por_ranking)
```

We see similar results as we saw for the maths dataset.

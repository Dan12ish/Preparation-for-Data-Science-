###Practice Program 10
A function which can perform all arithmetic operation on 2 numbers.

```python
def all(a,b):
  print('addition:             a+b = ', a+b)
  print('subtraction:          a-b = ', a-b)
  print('multiplication:       a*b = ', a*b)
  print('division:             a/b = ', a/b)
  print('a raise to the power: a**b = ', a**b)
  print('modulus:              a%b = ', a%b)
  print('floor division:       a//b = ', a//b)
```

```python
all(2,3)
```
